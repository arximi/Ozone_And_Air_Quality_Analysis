```python
#Import required library for visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Distribution of ozone
sns.histplot(ozone_cleaned["Daily Max 8-hour Ozone Concentration"], bins=50, kde=True)
plt.title('Distribution of Daily Max 8-hour Ozone Concentration')
plt.xlabel('Ozone Concentration (ppm)')
plt.ylabel('Frequency')
plt.show()
```

```python
ozone_cleaned['Month'] = ozone_cleaned['Date'].dt.month
sns.boxplot(x='Month', y='Daily Max 8-hour Ozone Concentration', data=ozone_cleaned)
```

```python
#Trend Analysis
ozone_cleaned.groupby('CBSA Name')['Daily Max 8-hour Ozone Concentration'].mean().sort_values(ascending=False).head(1000).plot(kind='barh',figsize=(6,8))
plt.title("CBSA vs. Ozone Conc.")
plt.xlabel("conc.")
plt.ylabel("CBSA")
plt.show()
```

```python
#Group by date to find average ozone concentration each day
daily_average= ozone_cleaned.groupby('Date')['Daily Max 8-hour Ozone Concentration'].mean().reset_index().head(500)
plt.figure(figsize=(24,8))
sns.lineplot(daily_average, x="Date", y="Daily Max 8-hour Ozone Concentration")
plt.title("Date vs. Ozone Conc.")
```

```python
#Group by county to find average ozone concentration by region
region_avg = ozone_cleaned.groupby('County')['Daily Max 8-hour Ozone Concentration'].mean().sort_values(ascending=False)
region_avg.plot(kind='bar', figsize=(12,5), title="Average Ozone by Region")
plt.ylabel("Ozone Conc.")
plt.xlabel("County")
plt.show()
```

```python
#Group by CBSA Name to find average ozone concentration by CBSA Name
region_avg = ozone_cleaned.groupby('CBSA Name')['Daily Max 8-hour Ozone Concentration'].mean().sort_values(ascending=False)
region_avg.plot(kind='bar', figsize=(12,5), title="Average Ozone by CBSA Name")
plt.ylabel("Ozone Conc.")
plt.xlabel("CBSA Name")
plt.show()
```

```python
#Ozone Level Trends
ozone_cleaned['day_of_week'] = ozone_cleaned['Date'].dt.day_name()
ozone_cleaned['is_weekend'] = ozone_cleaned['day_of_week'].isin(['Saturday', 'Sunday'])
```

```python
ozone_cleaned.shape
```

```python
sns.boxplot(data=ozone_cleaned, x='is_weekend', y='Daily Max 8-hour Ozone Concentration')
plt.title('Ozone Levels: Weekend vs. Weekday')
```

```python
#Extract month for seasonal trend
ozone_cleaned['Month'] = ozone_cleaned['Date'].dt.month
```

```python
#Add season column
def get_season(Month):
    if Month in [12, 1, 2]:
        return 'Winter'
    elif Month in [3, 4, 5]:
        return 'Spring'
    elif Month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

ozone_cleaned['Season'] = ozone_cleaned['Month'].apply(get_season)
```

```python
ozone_cleaned.shape
```

```python
#Boxplot by month
plt.figure(figsize=(12, 6))
sns.boxplot(x='Month', y='Daily Max 8-hour Ozone Concentration', data=ozone_cleaned)
plt.title('Ozone Concentration by Month')
plt.xlabel('Month')
plt.ylabel('Ozone Concentration (ppm)')
plt.show()
```
